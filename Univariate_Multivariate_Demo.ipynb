```python
import seaborn as sns
import pandas as pd #importing libraries as best practice
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
diamonds = sns.load_dataset('diamonds')
```

```python
diamonds.head()
```

```python
sns.histplot(diamonds,                 #Selects the whole data set to display as a histogram
             color='#F15B2a',          #Selects the color of the plot
             bins=50,                  #Selects the max amount of bins allowable
             edgecolor='#F15B2a',      #Selects the edge color of the plot
             linewidth=2)              #Selects the Line Width

plt.show()                             #Displays the Plot
```

```python

#Univariate Analysis
```

```python
sns.histplot(diamonds['carat'],      #Selects the singular variable 'carat'
             color='#F15B2a',          #Selects the color of the plot
             bins=50,                  #Selects the max amount of bins allowable
             edgecolor='#F15B2a',      #Selects the edge color of the plot
             linewidth=2)              #Selects the Line Width

plt.show()                             #Displays the Plot
```

```python
carat_mean = np.mean(diamonds['carat'])     #numpy mean
carat_mode = diamonds['carat'].mode()       #pandas mode
carat_median = diamonds['carat'].median()   #pandas median
carat_std = diamonds['carat'].std()         #pandas standard deviation
carat_var = diamonds['carat'].var()         #pandas variance
carat_skew = diamonds['carat'].skew()       #pandas skew
carat_kurt = diamonds['carat'].kurt()       #pandas kurt

print(carat_mean)
print(carat_mode)
print(carat_median)
print(carat_std)
print(carat_var)
print(carat_skew)
print(carat_kurt)
```

```python
diamonds.describe() #More efficient way to do the above functions
```

```python
sns.set(color_codes=True)

sns.kdeplot(diamonds['carat'], fill=True)

plt.axvline(diamonds['carat'].mean(), 0, 1)
plt.axvline(diamonds['carat'].median(), 0, 1, color='r')
```
